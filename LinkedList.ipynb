```python
class Node:
  def __init__(self, data=None, next=None):
    self.data=data
    self.next=next

class LinkedList:
  def __init__(self):
    self.head=None

  def insert_at_beginning(self,data):
    node=Node(data,self.head)
    self.head=node

  def print(self):
    itrstring=""
    if self.head is None:
      print("LL is empty")
      return
    itr=self.head
    while itr:
      itrstring+=str(itr.data)+ '-->'
      itr=itr.next
    print(itrstring)

  def insert_at_end(self,data):
    if self.head is None:
      self.head=Node(data,None)
      return
    itr=self.head
    while itr.next:
      itr=itr.next
    itr.next=Node(data, None)

  def insert_values(self, data_list):
    self.head=None
    for i in data_list:
      self.insert_at_end(i)

  def length_list(self):
    count=0
    itr=self.head
    while itr:
      count+=1
      itr=itr.next
    return count

  def remove_at(self, index):
    if index<0 or index>=self.length_list():
      print("Invalid index")
      return
    if index==0:
      self.head=self.head.next
    itr=self.head
    count=0
    while itr:
      if count==index-1:
        itr.next=itr.next.next
        break
      itr=itr.next
      count+=1

  def insert_at(self,data,index):
    if index<0 or index>=self.length_list()+1:
      print("Invalid Index")
    if index==0:
      self.insert_at_beginning(data)
      return
    itr=self.head
    count=0
    while itr:
      if count==index-1:
        node=Node(data,itr.next.next)
        itr.next=node
        break
      itr=itr.next
      count+=1

  def insert_after_value(self, data, value):
    if self.length_list()==0:
      raise Exception("empty LL inserting at beginning")
      self.insert_at_beginning(data)
      return
    itr=self.head
    while itr:
      if itr.data==value:
        node=Node(data,itr.next)
        itr.next=node
        break
      itr=itr.next

  def remove_by_value(self,value):
    if self.length_list()==0:
      raise Exception("empty LL")
      return
    itr=self.head
    while itr:
      if itr.next.data==value:
        itr.next=itr.next.next
        break
      itr=itr.next



ll=LinkedList()
ll.insert_at_end(235)
ll.insert_at_beginning(65)
ll.insert_at_beginning(66)
ll.insert_at_beginning(655)
ll.insert_at_end(235)
ll.insert_values([2,3,4,5,6,7,89])
ll.remove_at(0)
ll.insert_at(12,2)
ll.insert_after_value(12,6)
ll.print()
ll.remove_by_value(12)
ll.remove_by_value(12)
ll.print()
print(ll.length_list())
```
